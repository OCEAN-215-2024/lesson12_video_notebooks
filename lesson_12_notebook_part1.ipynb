```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
```

## Line plot examples

```python
# Create some data
x = np.array([1,2,3,4,5,6])
y = x**2

# Make a figure
fig = plt.figure()

# Plot the data 
plt.plot(x, y, c='r', ls='-', marker='o', ms=5, lw=2)

# Format the figure
plt.xlabel('x')
plt.ylabel('y')
plt.title('Example line plot');
```

```python
# Create some data
x = np.array([1,2,3,4,5,6])
y = x**2

# Make a figure
fig = plt.figure()

# Plot the data 
plt.plot(x, y, c='b', ls='--', marker='s', ms=5, lw=0.5)

# Format the figure
plt.xlabel('x')
plt.ylabel('y')
plt.title('Example line plot');
```

## Scatter plot examples

```python
# Make some data
x = np.array([1, 2, 3, 4, 5, 6])
y = x**2

# Create a figure
fig = plt.figure()

# Make a plot 
scat = plt.scatter(x,y, s=40, c='r', marker='o', ls='-', lw=1, edgecolor='k')

# Format the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Example scatter plot');
```

```python
# Make some data
x = np.array([1, 2, 3, 4, 5, 6])
y = x**2
z = x**3

# Create a figure
fig = plt.figure()
ax = plt.gca()

# Make a plot 
scat = plt.scatter(x, y, s=100, c=z, marker='o', ls='--', lw=1, edgecolor='k')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Example plot')

c = plt.colorbar(scat, ax=ax)
c.set_label('x**3')
```

```python
# Make some data
x = np.array([1, 2, 3, 4, 5, 6])
y = x**2
z = x**3

# Create a figure
fig = plt.figure()
ax = plt.gca()

# Make a plot 
scat = plt.scatter(x,y, s=100, c=z, marker='o', ls='--', lw=1, edgecolor='k', cmap='afmhot')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Example plot')

c = plt.colorbar(scat, ax=ax)
c.set_label('x**3')
```

## Histogram example

```python
# Change the filepath to wherever your data file is
filepath = 'data/woa18_temp.nc'

# Read the data - times are in a bad format, so don't parse them
temp = xr.open_dataset(filepath, decode_times=False)
t_data = temp['t_an'].values.flatten()
```

```python
# Create a figure
fig = plt.figure(figsize=(7, 4))

# Plot a histogram
h = plt.hist(t_data, bins=100, rwidth=0.6, color='b',edgecolor='k', alpha=0.5)

# Format the figure
plt.xlabel('Temperature ˚C')
plt.ylabel('Count')
plt.title('WOA Global Temperature Data');
```

```python
# Create a figure
fig = plt.figure(figsize=(7, 4))

# Plot a histogram
h = plt.hist(t_data, bins=100, rwidth=0.6, color='b',edgecolor='k', alpha=0.9)

# Format the figure
plt.xlabel('Temperature ˚C')
plt.ylabel('Count')
plt.title('WOA Global Temperature Data');
```

```python
# Create a figure
fig = plt.figure(figsize=(7,4))

# Plot a histogram
h = plt.hist(t_data, bins=range(0, 31), rwidth=0.6, color='b', edgecolor='k', alpha=0.9)

# Format the figure
plt.xlabel('Temperature ˚C')
plt.ylabel('Count')
plt.title('WOA Global Temperature Data');
```
